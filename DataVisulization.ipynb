```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=[20,25,35,55]
plt.plot(x,y)
plt.title("Line chart")
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.show()
```

```python
#Bar chart: it displays categorical data using rect bars whose lengths are proportional to the values.
#matplotlib.pyplot.bar(x, height)

import matplotlib.pyplot as plt
x=['Thu', 'Fri', 'Sat', 'Sun']
y=[170,120,250,190]

plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Histogram -> it shows the distribution of data by grouping values into bins
#matplotlib.pyplot.hist(x, bins=None)
#x=input data
#y=bins (number of bins(interval) to group data)
import matplotlib.pyplot as plt

x = [7, 8, 9, 10, 10, 12, 12, 12, 13, 14, 14, 15, 16, 16, 17, 18, 18, 19, 20, 20,
     21, 22, 23, 24, 25, 25, 26, 28, 30, 32, 35, 36, 38, 40, 42, 44, 48, 50]

plt.hist(x, bins=11, color='steelblue')
plt.title("Histogram")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter plot:   it is used to relatonship between variables
#matplotlib.pyplot.scatter(x,y)
#a scatter plot to visualize the relationship betwwen days and bill
import matplotlib.pyplot as plt

x = ['Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri', 'Sat', 'Sun']
y = [170, 120, 250, 190, 160, 130, 240, 200]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Pie chart is a circular chart used to show data as proportions or percentages.
#matplotlib.pyplot.pie(x, labels=None, autopct=None)
#x=data values for pie slices, labels, autopact-format to display percentage

import matplotlib.pyplot as plt
import pandas as pd
cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

plt.pie(data, labels=cars)
plt.title(" Pie Chart")
plt.show()
```

```python
#box plot -> shows how data is spread out, helps to spot outliers,it displays min, max, median
#matplotlib.pyplot.boxplot(x, notch=False, vert=True)
#x=data
#notch: if true, draws a notch to show interval around
import matplotlib.pyplot as plt
data = [ [10, 12, 14, 15, 18, 20, 22],
         [8, 9, 11, 13, 17, 19, 21],
         [14, 16, 18, 20, 23, 25, 27] ]

plt.boxplot(data)
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```

```python
#heatmap-> its a graphical representation of data where values are shown as colore, it helps to visulaize patters, correlations
#matplotlib.pyplot.imshow(X, cmap='viridis')
#x=2d array
#cmap-sets the color map
#np.random.seed(0): ensure some random values 
#np.random.rand(10,10): generates 10*10 random numbers b/w 0 and 1

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data = np.random.rand(10, 10)

plt.imshow(data, cmap='viridis', interpolation='nearest')

plt.colorbar()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Heatmap')
plt.show()
```

```python
Color: Change the color of the line
Linewidth: Adjust the width of the line
Marker: Change the style of plotted points
Markersize: Change the size of the markers
Linestyle: Define the style of the line like solid, dashed, etc.
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y, color='green', linewidth=3, marker='o', markersize=15, linestyle='--')

plt.title("Customizing Line Chart")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
Color: Fill color of the bars
Edgecolor: Color of the bar edges
Linewidth: Thickness of the edges
Width: Width of each bar
```

```python
import matplotlib.pyplot as plt

x = ['Thur', 'Fri', 'Sat', 'Sun']
y = [170, 120, 250, 190]

plt.bar(x, y, color='green', edgecolor='black', linewidth=2)

plt.title("Customizing Bar Chart")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Line plot:  A line plot shows the relationship between two numeric variables
#sns.lineplot(x=None, y=None, data=None)
import pandas as pd
import matplotlib.pyplot as plt
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df=pd.DataFrame(data)
plt.plot(df.index, df['Age'])
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age line plot')
plt.show()
```
